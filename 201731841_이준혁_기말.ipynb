## 1. 오픈 데이터 설명

* 1.1. 데이터셋[1)](https://www.bigdata-region.kr/#/dataset/9a362cbc-33a7-49aa-a0b0-eeefed39bd12) 찾은 배경    
대학 오기 전까지 강원도에 살다가 20살부터 홀로 경기도에 살면서 경기도는 다른 도에 비해서 청년에 대한 혜택[2)](https://www.jobaba.net/thema/exprcDtl.do?seq=3875&cntntsSeCd=02&shareUsr=)이 다양하다는 것을 알 게 되었다. 청년 기본소득, 청소년 교통비 지원, 청년 면접 수당, 청년 노동자 통장까지 찾아본다면 받을 수 있는 혜택이 많다. 빅데이터 과제를 찾기 위해서 여러 데이터 포털 사이트를 서칭하다가 경기청년 포털에 있는 평소 관심 있던 사업인 경기도 청년통장 선발정보에 대한 데이터셋을 찾게 되었다. 해당 산업은 경기도 거주 기준중위소득 100% 이하 가구의 청년이 매달 10만원을 저축하면 2년 후 경기도 예산으로 580만원이 적립되는 통장으로, 경기도 청년 노동자의 노동의지·취업의지를 고취시키고 금융역량을 강화하여 미래에 대한 대응력을 키우려는 목적을 가지고 있다.(총지급액은 580만원으로, 480만원 현금과 100만원 지역화폐 지급)

* 1.2. 다루고자 하는 내용  
경기도 청년 통장 선발 정보 파일은 많이 다뤄진 데이터셋은 아니었지만 기본적인 시각화 분석에 대한 자료가 있었고, 뻔한 시각화 및 분석을 할 것 같아서 포기했었다. 그러던 와중에 경기도 남북부에 대한 유튜브[3)](https://youtu.be/rpecqCS8LgM)를 보게 되었다. 경기도는 남북부가 불균형으로 인해서 교육, 정치 등 여러 가지 불균형 문제를 안고 있는데 그중에서도 북부가 수도권이라는 이유로 개발에 있어서도 다른 지방보다 불리하기 때문에 남북부를 분리해야 한다는 내용이었다. 이걸 보면서 이 데이터셋에 있는 변수를 활용해서 경기 남북부로 데이터셋을 나눠서 분석을 진행해 보면 좋겠다는 영감을 얻었다. 남북부로 나눠서 이 데이터셋을 바라보고 최종적으로는 실제 KOSIS에 나와있는 20~30대 인원을 고려해서 남북부 중에 어디가 더 해당 사업의 혜택을 받고 있는 지도 알아볼 것이다. 일반적으로 생각해 봤을 때 북부보다 남부가 더 잘 살기 때문에 북부의 인원이 더 혜택을 볼 것이라고 생각하지만 기준 중위소득 100% 이하 가구의 청년에 해당하는 인원들이기 때문에 일반적으로 생각하는 것과는 다른 결과가 나올 수 도 있다고 본다.

## 6. 데이터 분석 및 결과

### 6.1. 데이터 분석

데이터 시각화 파트에서 박스플롯을 통하여 남부, 북부의 평가총점이 비슷하게 나온 것을 알 수 있었다.  
그렇다면 이제 통계 분석을 남부, 북부간의 평가총점의 평균 차이가 있는지 알아보겠다.

$Ho:$ 경기도 일자리 청년통장 혜택을 받는 남북부 집단간의 평가총점의 평균 차이는 없다.  
$H1:$ 경기도 일자리 청년통장 혜택을 받는 남북부 집단간의 평가총점의 평균 차이는 있다.  

분석을 위해서 scipy.stat 라이브러리를 불러왔다.  
남부의 평균은 73.33, 북부의 평균은 75.23으로 비슷해보인다.  

````python
import scipy.stats
````

````python
a = jb_pb.loc[jb_pb['남북부'] == "남부", "평가총점"]
a.mean()
````

````python
b = jb_pb.loc[jb_pb['남북부'] == "북부", "평가총점"]
b.mean()
````

t-test를 진행하기에 앞서서 등분산 테스트를 진행했다.  
가설을 다음과 같다.  

$Ho:$ 남북부간의 분산은 같다.  
$H1:$ 남북부간의 분산은 같지 않다. 

levene 검사를 통해서 p-value값 < 0.05 이므로 귀무가설(H0)를 기각하여 집단간의 분산은 같지 않다는 것은 통계적으로 유의하다.  
따라서 t-test에서 equal_var=False로 두고 이분산이라고 가정하고 진행했고, p-value값 < 0.05이므로 귀무가설(H0)을 기각했다.  
시각적으로 봤을 때는 두 집단간의 차이가 없어 보였지만 통계 분석을 통해서는 두 집단간의 유의미한 차이가 있는 것으로 나왔다.

````python
scipy.stats.levene(a, b)
````

````python
scipy.stats.ttest_ind(a, b, equal_var=False)
````

## 8. 데이터 분석 수업을 마감하며 회고

지금까지는 학부 통계 수업을 들으면 주어진 예제에 대해서 배웠던 부분을 적용하기만 해서 별로 어려운 부분도 없었고 쉽게 해결이 되었다. 하지만 빅데이터 수업의 기말고사 대체과제는 온전히 혼자 힘으로 분석할 데이터셋을 찾고 데이터 전처리과정을 진행해야 해서 ‘어떤 식으로 진행해나가야 할까?’라는 어려움이 있었다. 또한, 해당 도메인에 관해서도 지식이 있어야 이 결측치를 어떤 식으로 처리해야 할지에 대한 판단도 내릴 수 있으므로 쉽지 않았다. 그리고 팀의 일부로서 특정 파트에 대한 분석을 진행하거나 기초적인 예제로 데이터를 다뤄본 경험은 있지만, 학부 4년 동안 나 혼자의 힘으로 데이터 분석 파이프라인에 맞춰서 분석을 진행한 경험이 없었다. 이 프로젝트를 통해서 데이터 분석 파이프라인에 대해서는 확실하게 기억이 남고 간다.

전체적으로 봤을 때 크게 성취한 부분은 이론적인 지식뿐만 아니라 프로그램을 다루는 능력인 것 같다. 먼저 마크다운은 간단하게 주석 용도로만 사용했지만, 이번 데이터 분석프로젝트를 통해서 사진, 수식 등의 코드들을 이해하며 정리할 수 있어서 마크다운이라는 것이 단순하게 주석 용도가 아닌 해당 분석의 파이프라인을 이해할 수 있게 해주는 중요한 작업임을 알 수 있었다. 실제 대학교 학부 수업에서는 프로그램을 배우더라도 예제데이터를 가지고 기본적인 함수를 적용해보고 이해하는 게 전부여서 어려움이 없었다. 하지만 이 수업에서 진행한 데이터 분석프로젝트에서는 데이터 소개 및 수집부터 시작해서 분석, 도출까지 하나의 파이프라인에 맞춰서 결과를 도출해나가는 것이기 때문에 막히는 부분이 발생했다. 하지만 그때마다 모르는 부분을 해결하기 위해서 구글링도 하고 책도 찾아보면서 해결해나갔다. 그로 인해서 Python을 다루는 능력이 전보다 많이 늘었다는 것을 느낄 수 있었다.

그리고 통계학과 학부 수업에서는 데이터 전처리과정에 대해서 중요하게 다루고 넘어간 수업이 없어서 머릿속으로 중요하다고는 생각은 했지만 직접 다뤄본 적은 없었다. 하지만 빅데이터 수업을 통해서는 데이터 클린징 작업 및 데이터 분석 파이프라인에 대해서 좀 더 고민해보고 중요하다는 것을 깨달을 수 있었다. 특히, 직접 공공데이터를 다뤄보니까 상상이상으로 데이터에 이상이 많다는 것을 알 수 있었다. 예를 들어 청년 통장 선발정보인데 40대와 100대인 데이터들도 있었고, 분석과는 전혀 상관 없는 쓸데없는 변수가 많아서 처리를 하느라 시각화 및 분석 자체보다는 이런 전처리 부분에서 시간이 가장 오래 걸린 것 같다. 이 수업을 통해서 얻은 데이터 클린징 경험들이 앞으로 캐글, 데이콘 같은 공모전에 나갈 때도 큰 도움이 될 것 같다.

분석에 있어서 가장 중요한 것은 데이터 전처리이다. <span style="font-size: 18px; font-weight: bold; color: red;">“Garbage in, garbage out”</span>이라는 말처럼 쓰레기를 넣어 봐야 쓰레기가 나오게 된다.  
이 말처럼 앞으로도 데이터 분석 파이프라인에 맞춰서 진행하면서 데이터 전처리과정에 대한 중요성을 가슴속에 새기고 데이터 분석을 할 것이다.
